# Plot the graph between Literacy and population density in 2015 year.

```python
df_literacy=df[df["Series Name"]=="Literacy rate, youth male (% of males ages 15-24)"]
df_literacy.head()
```

```python
df_population=df[df["Series Name"]=="Population density (people per sq. km of land area)"]
df_population.head()
```

```python
df_population.drop("Series Name",axis=1,inplace=True)
df_literacy.drop("Series Name",axis=1,inplace=True)
```

```python
X1=df_population["Country Name"]
```

```python
Y1=df_population['2015 [YR2015]']
```

```python
X2=df_literacy["Country Name"]
```

```python
Y2=df_literacy["2015 [YR2015]"]
```

```python
import matplotlib.pyplot as plt

plt.bar(X1, Y1)
plt.bar(X2,Y2)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Literacy vs Population')

# Display the plot
plt.show()
```
